#### a) Visualização de intervalo de dados em modo texto: a partir de entradas do usuário, sua implementação deve permitir a visualização dos dados que foram carregados do arquivo. O usuário deve informar o período que quer ver, ou seja, deve indicar o mês e ano iniciais, bem como o mês e ano finais que deseja visualizar os dados. Permita também que o usuário informe se quer ver 1) todos os dados, 2) apenas os de precipitação, 3) apenas os de temperatura, ou 4) apenas os de umidade e vento para o período informado. 

```python
import os
import pandas as pd

def subtitulo(prompt):
    os.system('cls' if os.name == 'nt' else 'clear')
    linha = '-' * (len(prompt))
    print(linha) 
    print(prompt)
    print(linha)
    print(' ')

def opcao_invalida():
    input('Opção inválida! Aperte qualquer tecla para voltar ao início.')
    main()

def voltar():
    os.system('cls' if os.name == 'nt' else 'clear')
    input('Opção fora dos valores requeridos, aperte qualquer tecla para voltar.')
    main()

def encerrar_app():
    subtitulo('Programa encerrado.')

def escolher_data(prompt):
    while True:
        try:
            data = input(prompt)
            data_convertida = pd.to_datetime(data, format='%d/%m/%Y')
            return data_convertida
        except ValueError:
            print("Formato inválido. Por favor, use dd/mm/aaaa.")

def exibir_opcoes():
    subtitulo('Como deseja visualizar os dados?')
    print("1. Todos os dados.")
    print("2. Apenas os dados de precipitação.")
    print("3. Apenas os dados de temperatura")
    print("4. Apenas os dados de umidade e velocidade relativa do vento")
    print("5. Sair.")

def all_data(df_filtrado):
    print(df_filtrado)

def precip_data(df_filtrado):
    print(df_filtrado[['data', 'precip']])

def temp_data(df_filtrado):
    print(df_filtrado[['data', 'maxima', 'minima', 'temp_media']])

def resto(df_filtrado):
    print(df_filtrado[['data', 'um_relativa', 'vel_vento']])

def escolher_opcoes(df_filtrado):
    '''Controle de escolha de opções do programa'''
    try:
        opcao_escolhida = int(input('Escolha uma opção: '))
        if opcao_escolhida == 1:
            all_data(df_filtrado)
        elif opcao_escolhida == 2:
            precip_data(df_filtrado)
        elif opcao_escolhida == 3:
            temp_data(df_filtrado)
        elif opcao_escolhida == 4:
            resto(df_filtrado)
        elif opcao_escolhida == 5:
            print('Saindo...')
            encerrar_app()
        else:
            opcao_invalida()
    except ValueError:
        opcao_invalida()

def main():
    os.system('cls' if os.name == 'nt' else 'clear')

    global df_clima
    df_clima = pd.DataFrame(df_clima)
    
    df_clima['data'] = pd.to_datetime(df_clima['data'], format='%d/%m/%Y')

    subtitulo('Intervalo de tempo que deseja visualizar')
    periodoinicial = escolher_data('Informe o valor do período inicial entre 1961 e 2016 para visualizar (no formato dd/mm/aaaa): ')
    periodofinal = escolher_data('Informe o valor do período final entre 1961 e 2016 para visualizar (no formato dd/mm/aaaa): ')

    df_filtrado = df_clima.loc[(df_clima['data'] >= periodoinicial) & (df_clima['data'] <= periodofinal)]

    exibir_opcoes()
    escolher_opcoes(df_filtrado)

if __name__ == '__main__':
    main()
```

#### b) Mês mais chuvoso

```python
def carregar_dados(file_path):
    df = pd.read_csv(file_path, parse_dates=['data'], dayfirst=True)
    return df

# Função para encontrar o mês mais chuvoso
def mes_mais_chuvoso(df):
    precip_por_mes = {} #Dicionário contendo as informações de precipitação

    # Agrupar os dados por mes/ano e somar as precipitações
    for index, row in df.iterrows():
        mes_ano = row['data'].strftime('%m/%Y')
        if mes_ano not in precip_por_mes:
            precip_por_mes[mes_ano] = 0
        precip_por_mes[mes_ano] += row['precip']

    # Encontrar o mes/ano com maior precipitação
    mes_chuvoso = max(precip_por_mes, key=precip_por_mes.get)
    maior_precip = precip_por_mes[mes_chuvoso]

    return mes_chuvoso, maior_precip, precip_por_mes

def grafico_precip(precip_por_mes):
    df_precip = pd.DataFrame(list(precip_por_mes.items()), columns=['mes/ano', 'Precipitação'])

    fig = px.bar(df_precip, x='mes/ano', y='Precipitação', title='Precipitação Mensal', labels={'mes/ano': 'mes/ano', 'Precipitação': 'Precipitação (mm)'})
    fig.update_layout(xaxis_tickangle=-45)
    fig.show()

def main():
    file_path = 'Anexo_Arquivo_Dados_Projeto_Logica_e_programacao_de_computadores.csv'
    df = carregar_dados(file_path)

    # Encontrar o mês mais chuvoso e a precipitação por mês
    mes_chuvoso, maior_precip, precip_por_mes = mes_mais_chuvoso(df)

    print(f"O mês mais chuvoso foi {mes_chuvoso}, com uma precipitação total de {maior_precip} mm.")

    grafico_precip(precip_por_mes) #Gráfico criado para melhor visualização dos dados

if __name__ == '__main__':
    main()
```

#### c) Média da temperatura mínima de um determinado mês (auge do inverno) nos últimos 11 anos (2006 a 2016);
#### d) Gráfico de barras (vertical ou horizontal) com as médias de temperatura mínima de um determinado mês nos últimos 11 anos;
#### e) Média geral da temperatura mínima de um determinado mês nos últimos 11 anos (2006 a 2016).

```python
import pandas as pd
import plotly.express as px

# Lista de meses válidos
meses_validos = ['janeiro', 'fevereiro', 'março', 'abril', 'maio', 'junho', 'julho', 'agosto', 'setembro', 'outubro', 'novembro', 'dezembro']

# Função para validar o mês informado pelo usuário
def validar_mes(mes):
    return mes.lower() in meses_validos

# Função para calcular a média da temperatura mínima de um mês específico nos últimos 11 anos (2006-2016)
def media_temp_min(df, mes):
    temp_min_media = {}
    mes_index = meses_validos.index(mes) + 1  # Índice do mês na lista (1-12)
    for ano in range(2006, 2017):
        # Filtrar os dados para o mês e ano específicos
        dados_mes_ano = df[(df['data'].dt.year == ano) & (df['data'].dt.month == mes_index)]
        
        # Calcular a média da temperatura mínima
        media_temp_minima = dados_mes_ano['minima'].mean()
        
        # Armazenar o resultado no dicionário
        temp_min_media[f"{mes.capitalize()}{ano}"] = media_temp_minima
    
    return temp_min_media

# Função para criar o gráfico de barras com as médias calculadas
def criar_grafico_barras(temp_min_media, mes):
    # Converter o dicionário para um DataFrame
    df_temp_min = pd.DataFrame(list(temp_min_media.items()), columns=['Mês/Ano', 'Média Temp Mínima'])
    
    # Criar o gráfico de barras
    fig = px.bar(df_temp_min, x='Mês/Ano', y='Média Temp Mínima', title=f'Média da Temperatura Mínima em {mes.capitalize()} (2006-2016)',
                 labels={'Mês/Ano': 'Mês/Ano', 'Média Temp Mínima': 'Média Temp Mínima (°C)'},
                 color='Média Temp Mínima', color_continuous_scale='tempo')
    
    # Atualizar layout do gráfico
    fig.update_layout(xaxis_tickangle=-45)
    fig.show()

# Função para calcular a média geral da temperatura mínima de um mês específico nos últimos 11 anos
def media_geral_temp_min(temp_min_media):
    total_temp = 0
    count = 0
    for valor in temp_min_media.values():
        if not pd.isna(valor):
            total_temp += valor
            count += 1
    media_geral = total_temp / count if count != 0 else float('nan')
    return media_geral

def main():
    # Carregar os dados do arquivo
    file_path = 'Anexo_Arquivo_Dados_Projeto_Logica_e_programacao_de_computadores.csv'
    df = pd.read_csv(file_path, parse_dates=['data'], dayfirst=True)
    
    # Remover linhas que contêm valores nulos
    df = df.dropna()
    
    # Pedir ao usuário para inserir o mês desejado
    mes = input("Informe o mês para calcular a média da temperatura mínima (ex: agosto): ").strip().lower()
    
    # Validar o mês
    if not validar_mes(mes):
        print("Mês inválido! Certifique-se de digitar o nome do mês corretamente.")
        return
    
    # Calcular a média da temperatura mínima para o mês informado
    medias = media_temp_min(df, mes)
    
    # Mostrar os resultados
    for chave, valor in medias.items():
        print(f"A média da temperatura mínima em {chave} foi de {valor:.2f}°C.")
    
    print("\nNota: Os dados vão até junho (06) de 2016, logo o restante dos dados serão mostrados como nulos (NaN).")

    # Calcular e exibir a média geral
    media_geral = media_geral_temp_min(medias)
    print(f"\nA média geral da temperatura mínima em {mes.capitalize()} de 2006 a 2016 foi de {media_geral:.2f}°C.")
    
    # Criar e mostrar o gráfico de barras
    criar_grafico_barras(medias, mes)

if __name__ == '__main__':
    main()
```
